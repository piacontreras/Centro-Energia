# GRÁFICO DE TORTA TECNOLOGÍA

```python
from pylab import *
#leyendo los archivos
df=pd.read_csv('bloques.csv',encoding='latin1',dtype={"Año":str})
df1=pd.read_csv('tecnologias.csv',encoding='latin1',dtype={'CenNom':str})
df2=pd.read_csv('plpcen.csv',dtype={'Hidro':str,'CenNom':str})
#borrar los espacios en blanco
df['Año']=df['Año'].str.strip()
df1['CenNom']=df1['CenNom'].str.strip()
df2['CenNom']=df2['CenNom'].str.strip()
df2['Hidro']=df2['Hidro'].str.strip()
#creando una tabla de plpcen y tecnologia
tabla0=pd.merge(df1, df2,on='CenNom')
#juntando la tabla anterior con bloques
tabla1=pd.merge(tabla0,df,on='Bloque')
#filtrando por tecnologia
tabla1_1=tabla1[(tabla1['Tecnolgía']=='Hidroelectrica')]
tabla1_2=tabla1[(tabla1['Tecnolgía']=='Mci')]
tabla1_3=tabla1[(tabla1['Tecnolgía']=='SI')]
tabla1_4=tabla1[(tabla1['Tecnolgía']=='Turbogas')]
tabla1_5=tabla1[(tabla1['Tecnolgía']=='Turbovapor')]

sum_hidro=tabla1_1['CenEgen'].sum()
sum_mci=tabla1_2['CenEgen'].sum()
sum_si=tabla1_3['CenEgen'].sum()
sum_tg=tabla1_4['CenEgen'].sum()
sum_tv=tabla1_5['CenEgen'].sum()

impr = ["Hidroelectrica", "Mci", "SI", "Turbogas","Turbovapor"]
vol = [sum_hidro, sum_mci, sum_si, sum_tg, sum_tv]
expl =(0, 0.05, 0, 0,0)
pie(vol, explode=expl, labels=impr, autopct='%1.1f%%', shadow=True)
plt.title("Tecnologia", bbox={"facecolor":"0.8", "pad":5})
plt.legend()
```
